### Importação dos dados



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja2.head()
```

```python
loja2.head()
```

# 1. Análise do faturamento


```python
faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()
print('Loja 1')
print(f'Faturamento Loja 1: R$ {faturamento_loja1}')
print()
print('Loja 2')
print(f'Faturamento Loja 2: R$ {faturamento_loja2}')
print()
print('Loja 3')
print(f'Faturamento Loja 3: R$ {faturamento_loja3}')
print()
print('Loja 4')
print(f'Faturamento Loja 4: R$ {faturamento_loja4}')
```

```python
# Dados
lojas = ['loja1','loja2','loja3','loja4']
faturamento = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]

# Gráfico de barras
plt.figure(figsize=(8, 6))
plt.bar(lojas, faturamento, color='mediumseagreen')
plt.title('Faturamento por Loja')
plt.xlabel('Lojas')
plt.ylabel('Faturamento (R$)')
plt.grid(axis='y', linestyle='--', alpha=0.7);
```

# 2. Vendas por Categoria


```python
print('LOJA 1')
categoria_preco_loja1 = loja[['Categoria do Produto','Preço']].groupby('Categoria do Produto').sum().sort_values('Preço',ascending=False)
print(categoria_preco_loja1)

print()
print('Loja 2')
categoria_preco_loja2 = loja2[['Categoria do Produto','Preço']].groupby('Categoria do Produto').sum().sort_values('Preço',ascending=False)
print(categoria_preco_loja2)

print()
print('Loja 3')
categoria_preco_loja3 = loja3[['Categoria do Produto','Preço']].groupby('Categoria do Produto').sum().sort_values('Preço',ascending=False)
print(categoria_preco_loja3)

print()
print('Loja 4')
categoria_preco_loja4 = loja4[['Categoria do Produto','Preço']].groupby('Categoria do Produto').sum().sort_values('Preço',ascending=False)
print(categoria_preco_loja4)
```

```python
# Dados
categorias = loja['Categoria do Produto'].unique().tolist()
print(categorias)
loja1 = categoria_preco_loja1['Preço'].unique().tolist()
loja2 = categoria_preco_loja2['Preço'].unique().tolist()
loja3 = categoria_preco_loja3['Preço'].unique().tolist()
loja4 = categoria_preco_loja4['Preço'].unique().tolist()
print(loja1)
print(loja2)
print(loja3)
print(loja4)
```

```python
# Categorias fixas
categorias = [
    'eletronicos', 'moveis', 'brinquedos', 'eletrodomesticos',
    'utilidades domesticas', 'esporte e lazer', 'livros', 'instrumentos musicais'
]

# Valores de vendas por loja
loja1 = [572659.23, 250178.11, 23993.78, 484913.36, 16931.48, 52387.55, 11713.92, 121731.69]
loja2 = [547773.71, 235234.68, 21262.21, 464758.13, 19662.86, 46326.09, 13454.35, 139987.03]
loja3 = [547699.92, 268095.56, 25869.2, 438984.89, 20080.21, 47456.1, 12664.7, 103174.45]
loja4 = [545966.86, 256705.65, 27015.67, 377680.65, 20098.83, 44465.65, 12427.77, 100136.5]

lojas = [loja1, loja2, loja3, loja4]
titulos = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

# Criar subgráficos
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
axs = axs.ravel()

for i in range(4):
    axs[i].bar(categorias, lojas[i], color='skyblue')
    axs[i].set_title(f'Vendas por Categoria - {titulos[i]}')
    axs[i].set_ylabel('Total em R$')
    axs[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

# 3. Média de Avaliação das Lojas

```python
loja2.head()
```

```python
media_avaliacao_loja1 = loja['Avaliação da compra'].mean()
print(f'Média da avaliação da satisfação dos clientes para a loja 1: {round(media_avaliacao_loja1,2)}')

media_avaliacao_loja2 = loja2['Avaliação da compra'].mean()
print(f'Média da avaliação da satisfação dos clientes para a loja 2: {round(media_avaliacao_loja2,2)}')

media_avaliacao_loja3 = loja3['Avaliação da compra'].mean()
print(f'Média da avaliação da satisfação dos clientes para a loja 3: {round(media_avaliacao_loja3,2)}')

media_avaliacao_loja4 = loja4['Avaliação da compra'].mean()
print(f'Média da avaliação da satisfação dos clientes para a loja 4: {round(media_avaliacao_loja4,2)}')

```

```python
avaliacoes = [media_avaliacao_loja1, media_avaliacao_loja2, media_avaliacao_loja3, media_avaliacao_loja4]

# Gráfico de linha
plt.figure(figsize=(8, 5))
plt.plot(lojas, avaliacoes, marker='o', linestyle='-', color='green')

# Personalizações
plt.title('Média de Satisfação dos Clientes por Loja')
plt.ylabel('Nota de Satisfação (0 a 5)')
plt.ylim(3.9, 4.1)
plt.grid(True)
plt.tight_layout()

plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
produtos_loja1 = loja[['Produto','Preço']].groupby('Produto',as_index=False).sum().sort_values(by='Preço',ascending=False)
produtos_loja2 = loja2[['Produto','Preço']].groupby('Produto',as_index=False).sum().sort_values(by='Preço',ascending=False)
produtos_loja3 = loja3[['Produto','Preço']].groupby('Produto',as_index=False).sum().sort_values(by='Preço',ascending=False)
produtos_loja4 = loja4[['Produto','Preço']].groupby('Produto',as_index=False).sum().sort_values(by='Preço',ascending=False)
```

```python
# Cria lista produtos
lista_produtos_loja1 = produtos_loja1['Produto'].tolist()
lista_produtos_loja2 = produtos_loja2['Produto'].tolist()
lista_produtos_loja3 = produtos_loja3['Produto'].tolist()
lista_produtos_loja4 = produtos_loja4['Produto'].tolist()

# Cria lista preços

lista_precos_loja1 = produtos_loja1['Preço'].tolist()
lista_precos_loja2 = produtos_loja2['Preço'].tolist()
lista_precos_loja3 = produtos_loja3['Preço'].tolist()
lista_precos_loja4 = produtos_loja4['Preço'].tolist()
```

```python
plt.figure(figsize=(15, 8))
plt.barh(lista_produtos_loja1, lista_precos_loja1, color='skyblue')
plt.xlabel('Preço (R$)')
plt.title('Gráfico de Barras: Preços dos Produtos Loja1')
plt.tight_layout()  # Ajusta o layout para evitar sobreposição
plt.show()
```

```python
indices = range(len(lista_produtos_loja2))

plt.figure(figsize=(15, 7))
plt.scatter(indices, lista_precos_loja2, color='red', alpha=0.6, s=50)  # s = tamanho dos pontos
plt.xlabel('Produtos (índices)')
plt.ylabel('Preço (R$)')
plt.title('Gráfico de Dispersão: Preços dos Produtos Loja2')
plt.grid(True, linestyle='--', alpha=0.5)

# Rotacionar os rótulos do eixo X para melhor visualização
plt.xticks(indices, lista_produtos_loja2, rotation=90, fontsize=8)
plt.tight_layout()  # Ajusta o layout para evitar sobreposição
plt.show()
```

```python
# Ordenar produtos por preço (do maior para o menor)
dados_ordenados = sorted(zip(lista_produtos_loja3, lista_precos_loja3), key=lambda x: x[1], reverse=True)
top_20_produtos = [produto for produto, preco in dados_ordenados[:20]]
top_20_precos = [preco for produto, preco in dados_ordenados[:20]]

plt.figure(figsize=(12, 6))
plt.step(top_20_produtos, top_20_precos, where='mid', color='teal', linewidth=2)
plt.title('Top 20 Produtos Mais Caros (Gráfico Escada) Loja 3')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Ordenar por preço e pegar os 15 mais caros
dados_ordenados = sorted(zip(lista_produtos_loja4, lista_precos_loja4), key=lambda x: x[1], reverse=True)[:15]
produtos_top = [p[0] for p in dados_ordenados]
precos_top = [p[1] for p in dados_ordenados]

plt.figure(figsize=(12, 6))
plt.plot(produtos_top, precos_top, marker='s', color='orange', linewidth=2)
plt.title('Top 15 Produtos Mais Caros Loja 4')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 5. Frete Médio por Loja


```python
print('Loja 1')
gasto_medio_frete_loja1 = round(loja['Frete'].mean(),2)
print(f'Em média a loja 1 gasta com frete R$ {gasto_medio_frete_loja1}')
print()
print('Loja 2')
gasto_medio_frete_loja2 = round(loja2['Frete'].mean(),2)
print(f'Em média a loja 2 gasta com frete R$ {gasto_medio_frete_loja2}')
print()
print('Loja 3')
gasto_medio_frete_loja3 = round(loja3['Frete'].mean(),2)
print(f'Em média a loja 3 gasta com frete R$ {gasto_medio_frete_loja3}')
print()
print('Loja 4')
gasto_medio_frete_loja4 = round(loja4['Frete'].mean(),2)
print(f'Em média a loja 4 gasta com frete R$ {gasto_medio_frete_loja4}')
```

```python
# Dados
lojas = ['loja1', 'loja2', 'loja3', 'loja4']
frete_lojas = [gasto_medio_frete_loja1,gasto_medio_frete_loja2,gasto_medio_frete_loja3,gasto_medio_frete_loja4]

# Criando o gráfico de barras
plt.figure(figsize=(8, 5))
plt.bar(lojas, frete_lojas, color='magenta')
plt.title('Frete por Loja')
plt.xlabel('Lojas')
plt.ylabel('Valor do Frete (R$)')
plt.grid(axis='y', linestyle='--',alpha=0.5)

# Exibindo o gráfico
plt.tight_layout()
plt.show()
```

## Relatório Final – Análise de Desempenho das Lojas do Senhor João
### Introdução
O objetivo desta análise é auxiliar o Senhor João em uma decisão importante: qual loja vender entre as quatro que atualmente administra. Para isso, foram avaliados diversos fatores que impactam diretamente a performance e a sustentabilidade de cada loja, como: faturamento, avaliação dos clientes, frete médio, além dos produtos e categorias mais e menos vendidos.
### Desenvolvimento
#### Faturamento Total
A primeira métrica observada foi o **faturamento total** das lojas. Nessa análise, a **Loja 4** apresentou o **menor faturamento**, o que já levanta um alerta quanto à sua performance comercial.
#### Satisfação dos Clientes
Analisando a **média das avaliações dos clientes**, temos os seguintes resultados:
- Loja 1: 3.98
- Loja 2: 4.04
- Loja 3: 4.05 (maior média)
- Loja 4: 4.00
  
Apesar de a **Loja 4** não estar com a pior avaliação, ela também não é destaque neste quesito. A **Loja 3** se sobressai, com a maior média de satisfação entre os clientes.
#### Frete Médio
A **Loja 4** tem o **menor custo médio de frete: R$ 31,28**, o que é um ponto positivo. Contudo, essa vantagem de logística não foi suficiente para alavancar suas vendas.
#### Produtos e Categorias
As categorias mais vendidas em todas as lojas envolvem **eletrônicos, eletrodomésticos e móveis**, enquanto as menos vendidas foram **livros, brinquedos, utilidades domésticas e artigos de lazer**.
Nos dados detalhados, foi possível observar uma tendência clara em todas as lojas: os **produtos tecnológicos e eletrodomésticos dominam o ranking dos mais vendidos**, enquanto itens como **livros, brinquedos e jogos** ocupam as últimas posições.
Isso demonstra uma possível **desconexão entre o mix de produtos menos vendidos e a demanda dos consumidores**, especialmente na Loja 4, que possui baixa performance de vendas mesmo mantendo os mesmos produtos mais populares.

### Conclusão e Recomendação
Após analisar todos os fatores — **faturamento, satisfação do cliente, frete médio e desempenho de produtos —, a recomendação é que o Senhor João venda a Loja 4**.
Apesar de apresentar **menor custo de frete**, a Loja 4 tem o **menor faturamento, não se destaca na avaliação dos clientes e não tem uma vantagem significativa no mix de produtos** em relação às outras lojas. Ou seja, seus pontos positivos não compensam os negativos.
As demais lojas apresentam desempenho mais equilibrado e potencial de crescimento, principalmente a **Loja 3**, que combina **boa avaliação, bom faturamento e frete competitivo**.
Portanto, com base em dados concretos e objetivos, a venda da Loja 4 representa a decisão mais estratégica e financeiramente saudável para o Senhor João.
